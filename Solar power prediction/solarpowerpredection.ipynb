```python
import matplotlib.pyplot as plt
```

```python
# plot distribution of power
plt.figure(figsize=(10, 6))
sns.histplot(df['generated_power_kw'],bins=30,kde=True)
plt.title('Distribution of Generated Power')
plt.xlabel('Generated Power (kW)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bivarate analysis
# scatter plot for target feature
plt.figure(figsize=(15, 30))
for i, column in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    plt.scatter(df[column], df['generated_power_kw'])
    plt.title(f'{column} vs Generated Power')
    plt.xlabel(column)
    plt.ylabel('Generated Power (kw)')
    plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(df.corr(),cmap='coolwarm', annot=True, fmt='.2f')
plt.show()
```

```python
# outlier analysis
plt.figure(figsize=(15, 30))
for i, column in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    sns.boxplot(df[column])
plt.show()
```
